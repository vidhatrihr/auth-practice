```python
def generate_token(length=10):
  chars = string.ascii_letters + string.digits
  return ''.join(secrets.choice(chars) for _ in range(length))


generate_token()
```

```python
class Todo:
  next_id = 0
  table = {}

  def __init__(self, text):
    self.id = Todo.next_id
    self.text = text

    Todo.next_id += 1
    Todo.table[self.id] = self

  def __repr__(self):
    return f"<Todo id={self.id}, text={self.text}>"

  def dictify(self):
    return {
        "id": self.id,
        "text": self.text
    }


todo0 = Todo("drink water")
todo1 = Todo("save plants")
todo2 = Todo("eat apples")
todo3 = Todo("eat mangos")

Todo.table
```
