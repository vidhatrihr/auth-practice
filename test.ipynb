```python
import string
import secrets

secrets.choice('apple')

chars = string.ascii_letters + string.digits
''.join(secrets.choice(chars) for _ in range(10))
```

```python
def generate_token(length=10):
  chars = string.ascii_letters + string.digits
  return ''.join(secrets.choice(chars) for _ in range(length))


generate_token(10)
```
